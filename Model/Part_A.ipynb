# Part A : Probability & Conceptual Foundation

## Q1. What is Conditional Probability ?

### Conditional probability is the probability of an event occurring given that another event has already occurred. It helps measure how one event affects the likelihood of another event. It is represented as P(A|B), which means the probability of event A given event B.

## Q2. Explain Bayes' Theorem and its importance in classification problems.

### Bayes' Theorem is a mathematical formula used to calculate the probability of an event based on prior knowledge and new evidence. It updates the probability of a class after observing the input features. It is the foundation of the Naive Bayes classifier and is widely used for classification tasks such as spam detection and medical diagnosis.

## Q3. What assumption does Naive Bayes Classifier make ?

### The Naive Bayes classifier assumes that all features are conditionally independent of each other given the class label. This means the presence or absence of one feature does not influence another feature within the same class.

## Q4. Explain the working principle of :

## A. K-Nearest Neighbors ( KNN )

### KNN is a distance-based supervised learning algorithm that classifies a new data point by finding its K nearest neighbors in the training dataset. It calculates the distance (commonly Euclidean distance) between data points and assigns the class that appears most frequently among the nearest neighbors.

## B. Support Vector Machine ( SVM )

### SVM is a supervised learning algorithm that finds the optimal hyperplane to separate different classes with the maximum possible margin. The data points closest to the hyperplane are called support vectors, which determine the decision boundary. For non-linearly separable data, SVM uses kernel functions such as the RBF kernel.

## Q5. Compare distance-based, probabilistic, and margin-based classifiers.

## Distance Based :

### Uses the distance between data points for classification.
### Example : KNN 
### Performance depends on the choice of distance metric and value of K.

## Probabilistic :

### Uses probability distributions to predict the most likely class.
### Example : Naive Bayes
### Assumes conditional independence among features.

## Margin-based :

### Uses an optimal decision boundary with the maximum margin between classes.
### Example : SVM 
### Works well for high-dimensional and complex datasets.
